```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/kaggle/input/preprocessed-textual-data/preprocessed_text_data.csv")
```

```python
df = data
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(inplace =True)
```

```python
df.drop("index",axis = 1,inplace=True)
```

```python
### Load the vectorizer file

import pickle

with open("/kaggle/input/vectorization-pickle-file/vectorizer.pkl","rb") as file:
    vectorizer = pickle.load(file)
```

```python
### Splitting the data into training and testing

from sklearn.model_selection import train_test_split
```

```python
X = df["text"]
y = df["label"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
### Applying vectorization to data

x_train = vectorizer.fit_transform(x_train)
```

```python
x_test = vectorizer.transform(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

```python
decision_tree = DecisionTreeClassifier()
svm = SVC()
k_neighbours = KNeighborsClassifier()
xgboost = XGBClassifier(eval_metrics = "logloss")
catboost = CatBoostClassifier(verbose = 0)
```

```python
## Decisison Tree
print("-----Decision Tree-----")
decision_tree.fit(x_train,y_train)
print("----Training Accuracy----")
y_train_pred = decision_tree.predict(x_train)
print("Accuracy is ",accuracy_score(y_train,y_train_pred))
print(classification_report(y_train,y_train_pred))
print("-----Testing Accuracy-----")
y_pred= decision_tree.predict(x_test)
print("Accuracy is ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
## SVC
print("-----SVC-----")
svm.fit(x_train,y_train)
print("----Training Accuracy----")
y_train_pred = svm.predict(x_train)
print("Accuracy is ",accuracy_score(y_train,y_train_pred))
print(classification_report(y_train,y_train_pred))
print("-----Testing Accuracy-----")
y_pred= svm.predict(x_test)
print("Accuracy is ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
## SVC
print("-----XGBClassifier-----")
xgboost.fit(x_train,y_train)
print("----Training Accuracy----")
y_train_pred = xgboost.predict(x_train)
print("Accuracy is ",accuracy_score(y_train,y_train_pred))
print(classification_report(y_train,y_train_pred))
print("-----Testing Accuracy-----")
y_pred= xgboost.predict(x_test)
print("Accuracy is ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
## CatBoostClassifier
print("-----CatBoostClassifier-----")
catboost.fit(x_train,y_train)
print("----Training Accuracy----")
y_train_pred = catboost.predict(x_train)
print("Accuracy is ",accuracy_score(y_train,y_train_pred))
print(classification_report(y_train,y_train_pred))
print("-----Testing Accuracy-----")
y_pred= catboost.predict(x_test)
print("Accuracy is ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
